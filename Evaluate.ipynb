```python
import pandas as pd
import numpy as np
from blist import sortedset
from pyjudy import JudyLIntObj

col_name = {0 : 'user_id', 1 : 'brand_id'}
predict_df = pd.read_csv("data/season1/recommend.direct.txt", header=None, delim_whitespace=True, encoding="gbk").rename(columns=col_name)
test_df = pd.read_csv("data/season1/test.csv", encoding="gbk")
buy_df = test_df[test_df.type == 1]

# 预测集与测试集中的用户
predict_users = predict_df.user_id.unique()
test_users = buy_df.user_id.unique()

# 预测集与测试集中的用户购买物品列表
predict_map = JudyLIntObj()
test_map = JudyLIntObj()

# 预测集按UID分组
predict_df['brand_id'] = predict_df['brand_id'].astype(str)
predict_gb = predict_df.groupby('user_id')
for key, group in predict_gb:
    user_id = int(key)
    brand_ids = []
    # 把字符串转换成整型列表
    for b_index in group.index:
        brand_ids.extend([int(x.strip(' ')) for x in group.ix[b_index]['brand_id'].split(',')])
    u_map = predict_map.get(user_id)
    
    # 存放物品列表
    try:
        predict_map[user_id].update(sortedset(brand_ids))
    except KeyError:
        predict_map[user_id] = sortedset(brand_ids)

# 测试集按UID分组
test_gb = buy_df.groupby('user_id')
for key, group in test_gb:
    user_id = int(key)
    brand_ids = []
    # 把字符串转换成整型列表
    for b_index in group.index:
        brand_ids.append(group.ix[b_index]['brand_id'])
    u_map = test_map.get(user_id)
    
    # 存放物品列表
    if not u_map:
        test_map[user_id] = sortedset(brand_ids)
    else:
        u_map.update(sortedset(brand_ids))

# 预测集与测试集有不相交的用户以空列表形式存放
for uid in sortedset(predict_map.keys()).difference(sortedset(test_map.keys())):
    test_map[uid] = sortedset([])

for uid in sortedset(test_map.keys()).difference(sortedset(predict_map.keys())):
    predict_map[uid] = sortedset([])

# 准确率
p_hit_brands = sum([len(predict_map[int(user_id)].intersection(test_map[int(user_id)])) for user_id in predict_users])
p_brands = sum([len(predict_map[int(user_id)]) for user_id in test_users])
precision = p_hit_brands / p_brands

# 召回率
r_hit_brands = sum([len(predict_map[int(user_id)].intersection(test_map[int(user_id)])) for user_id in test_users])
r_brands = sum([len(test_map[int(user_id)]) for user_id in test_users])
recall = r_hit_brands / r_brands

score = (2 * precision * recall) / (precision + recall)

print "准确率: %f\n召回率: %f\n总得分: %f" % (precision, recall, score)
```
