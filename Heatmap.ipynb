```python
import pandas as pd
import scipy.sparse as sps
import numpy as np
import matplotlib.pyplot as plt
from itertools import izip

rate_mt = sps.lil_matrix((100, 100), dtype='float32')
for x, y in izip(np.random.randint(0, 100, 5000), np.random.randint(0, 100, 5000)):
    rate_mt[x, y] = np.random.random()
```

```python
# 分布图
import matplotlib.pylab as pl
import matplotlib.colors as mcolors
pl.spy(rate_mt.todense(), aspect='auto', cmap=mcolors.ListedColormap(['c', 'r']), alpha=0.8)
fig = plt.gcf()
fig.set_size_inches(20,20)
pl.show()
```

```python
# 热度图
y, x = np.mgrid[slice(0, rate_mt.shape[0] + 1), slice(0, rate_mt.shape[1] + 1)]
fig, ax = plt.subplots()
plt.pcolormesh(x, y, rate_mt.toarray(), cmap=plt.cm.Blues, alpha=0.8)
plt.axis([x.min(), x.max(), y.min(), y.max()])
plt.colorbar()
ax.xaxis.tick_top()
ax.invert_yaxis()
plt.show()
```

```python
# 热度图
fig, ax = plt.subplots()
heatmap = ax.pcolor(pd.DataFrame(rate_mt.todense()), cmap=plt.cm.Blues, alpha=0.8)

# 设置图片的格式
fig = plt.gcf()
fig.set_size_inches(20,20)

# 去掉图片边框
ax.set_frame_on(False)

# 设置坐标的刻度
ax.set_yticks(np.arange(rate_mt.shape[0])+0.5, minor=False)
ax.set_xticks(np.arange(rate_mt.shape[1])+0.5, minor=False)

# 把横轴的刻度置顶
ax.xaxis.tick_top()
ax.invert_yaxis()

# 设置坐标刻度的标签
ax.set_yticklabels(xrange(rate_mt.shape[0]), minor=False)
ax.set_xticklabels(xrange(rate_mt.shape[1]), minor=False)

# 旋转横轴刻度标签
plt.xticks(rotation=90)

ax.grid(False)

# 去掉所有的刻度
ax = plt.gca()
for t in ax.xaxis.get_major_ticks(): 
    t.tick1On = False
    t.tick2On = False
for t in ax.yaxis.get_major_ticks(): 
    t.tick1On = False
    t.tick2On = False
```
