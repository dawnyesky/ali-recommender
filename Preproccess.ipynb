```python
from __future__ import division
import pandas as pd
import numpy as np
import scipy as sp
import matplotlib.pylab as pl
import matplotlib.pyplot as plt
from scipy.sparse import dok_matrix, lil_matrix, csr_matrix
from pyjudy import JudyLIntInt, JudyLIntObj, JudySLObj

def to_timestamp(dt_unicodes):
    return [pd.Timestamp("2013-%i-%i" % (int(list(dt_unicode)[0]), int(list(dt_unicode)[2]))) for dt_unicode in dt_unicodes]
df = pd.read_csv("data/season1/train.csv", parse_dates=['visit_datetime'], date_parser=to_timestamp, encoding="gbk")

#df.sort(columns='visit_datetime', ascending=True, inplace=True)
df.describe()
```

```python
#print "user_id:\n%s\n\n\nbrand_id:\n%s\n\n\ntype:\n%s\n\n" % (df['user_id'].value_counts(), df['brand_id'].value_counts(), df['type'].value_counts())
```

```python
# 将type属性按照属性值转换成位串，并重命名
columns =  np.array(df['type'].unique().tolist())
columns.sort()
col_map = {0 : 'view', 1 : 'buy', 2 : 'favor', 3 : 'cart'}
def to_type_array(type_row):
    type_array = pd.Series(data=0, index=columns, dtype='int32')
    type_array[type_row['type']] = 1
    return type_array
type_df = pd.DataFrame({'type': df['type']})
type_array_df = type_df.apply(to_type_array, axis=1).rename(columns = col_map)
```

```python
# 将type位串替换原来的type属性
new_df = pd.merge(df, type_array_df, left_index=True, right_index=True)
del new_df['type']
```

```python
# 筛选时间区间
#period_df = new_df[new_df.visit_datetime.between(pd.Timestamp("2013-06-01"), pd.Timestamp("2013-08-31"))]
#new_df['view'].value_counts()
```

```python
# 按照user-brand分组
gb = new_df.groupby(['user_id', 'brand_id'])

def wrap_key(digits):
    digits_str = [str(x) for x in digits]
    return ",".join(digits_str)

# 对UID和BID重新编号并建立相应索引
uid_vc = df['user_id'].value_counts()
uid_map = JudyLIntInt()
uid_rmap = JudyLIntInt()
for uid in xrange(len(uid_vc)):
    uid_map.Ins(int(uid_vc.index[uid]), uid)
    uid_rmap.Ins(uid, int(uid_vc.index[uid]))
    
bid_vc = df['brand_id'].value_counts()
bid_map = JudyLIntInt()
bid_rmap = JudyLIntInt()
for bid in xrange(len(bid_vc)):
    bid_map.Ins(int(bid_vc.index[bid]), bid)
    bid_rmap.Ins(bid, int(bid_vc.index[bid]))
```

```python
# 线性回归训练集
reg_cols = ['user_id',  'brand_id', 'view', 'cart', 'favor', 'buy', 'visit_datetime']
linreg_dataset_list = []

# 逻辑回归训练集
logreg_dataset_list = []

# 建立UID-BID矩阵，每个元素存储历史行为记录
df_map = JudySLObj()
#event_map = JudySLObj()
view_mt = dok_matrix((len(uid_map), len(bid_map)), dtype='int32')
buy_mt = dok_matrix((len(uid_map), len(bid_map)), dtype='int32')
cart_mt = dok_matrix((len(uid_map), len(bid_map)), dtype='int32')
favor_mt = dok_matrix((len(uid_map), len(bid_map)), dtype='int32')
ub_event_map = JudyLIntObj()
b_event_map = JudyLIntObj()
for key, group in gb:
    # 计算索引
    index = ([uid_map[int(key[0])], bid_map[int(key[1])]])
    index_str = wrap_key([uid_map[int(key[0])], bid_map[int(key[1])]])
    
    # 基础统计
    view_mt[index] += group.view.sum()
    buy_mt[index] += group.buy.sum()
    cart_mt[index] += group.cart.sum()
    favor_mt[index] += group.favor.sum()
    
    # 暂存
    ub_df = group.set_index('visit_datetime')
    df_map[index_str] = ub_df
    
    # 线性回归分析预处理
    group_df = group.set_index('visit_datetime').ix[:, ['view', 'buy', 'favor', 'cart']].resample('15D', how='sum', closed={'right'}, loffset='15D').dropna().reset_index()
    group_df[['view', 'buy', 'favor', 'cart']] = group_df[['view', 'buy', 'favor', 'cart']].astype('float32')
    group_df['user_id'] = key[0]
    group_df['brand_id'] = key[1]
    group_df = group_df.reindex_axis(reg_cols, axis=1)
    linreg_dataset_list.append(group_df)
    
    # 逻辑回归分析预处理
    group_df = group.set_index('visit_datetime').ix[:, ['view', 'buy', 'favor', 'cart']].resample('1D', how='sum').dropna()
    buy_indices = group_df[group_df.buy > 0].index
    prv_index = group_df.index[0]
    for buy_index in buy_indices:
        if not buy_index.freq:
            buy_index = pd.DatetimeIndex([buy_index], freq='1D')[0]
        interval_days = "%iD" % max((buy_index - prv_index).days, 1)
        interval_df = group_df.ix[prv_index:buy_index].resample(interval_days, how='sum', closed={'right'}, loffset=interval_days).dropna().reset_index()
        interval_df[['view', 'buy', 'favor', 'cart']] = interval_df[['view', 'buy', 'favor', 'cart']].astype('float32')
        interval_df['user_id'] = key[0]
        interval_df['brand_id'] = key[1]
        interval_df['buy'] = 1
        interval_df = interval_df.reindex_axis(reg_cols, axis=1)
        logreg_dataset_list.append(interval_df)
        prv_index = buy_index + 1
    if (len(buy_indices) == 0 or prv_index < group_df.tail(1).index[0]):
        group_df = group_df.ix[prv_index:]
        last_timestamp = group_df.tail(1).index[0]
        group_df['key'] = 0
        interval_df = group_df.groupby(group_df.key).sum().astype('float32')
        interval_df.reset_index(drop=True, inplace=True)
        del interval_df['key']
        interval_df['user_id'] = key[0]
        interval_df['brand_id'] = key[1]
        interval_df['visit_datetime'] = last_timestamp
        interval_df = interval_df.reindex_axis(reg_cols, axis=1)
        logreg_dataset_list.append(interval_df)
    
    # 自定义度量
    group_df = group.set_index('visit_datetime').ix[:, ['view', 'buy', 'favor', 'cart']].resample('3D', how='sum', closed={'right'}, loffset=interval_days).dropna()
    group_df['done'] = pd.Series(data=0, index=group_df.index)

    # 统计UID-BID的各种事件次数，e[0]: 光看不买但浏览次数大于5；e[1]: 光看不买且浏览次数小于5；e[2]: 买后喜欢，回头客；e[3]: 买后不喜欢，一次性(0/1)；e[4]: 浏览后犹豫很久才购买；e[5]: 浏览后放入购物车却没有购买；e[6]: 收藏(0/1)
    e = np.zeros((1, 7), dtype='int32')[0]

    # 购买事件
    buy_indices = group_df[group_df.buy > 0].index
    if (len(buy_indices) > 0):
        prv_index = buy_indices[0]
    for buy_index in buy_indices:
        if not buy_index.freq:
            buy_index = pd.DatetimeIndex([buy_index], freq='3D')[0]
        # 购买前6天内的浏览不算
        group_df.ix[buy_index - 2:buy_index, 'done'] = 1
        # 6天前到上一次购买之间有浏览记录
        backward = group_df.ix[prv_index:buy_index - 2]
        if (len(backward[backward.view > 0].index) == len(backward.index)):
            e[4] += 1
        else:
            # 购买后6天内的浏览不算
            forward = group_df.ix[buy_index:buy_index + 2]
            group_df.ix[forward.index, 'done'] = 1
            # 购买后有再次购买或再次浏览
            if ((buy_index - prv_index).days > 0 or forward.view.sum() > 0):
                e[2] += 1
            elif ((pd.Timestamp('now') - buy_index).days > 30):
                e[3] += 1
        prv_index = buy_index

    # 浏览事件
    for view_index in group_df[group_df.view > 0].index:
        if (group_df.ix[view_index]['done'] == 1):
            continue
        if (group_df.ix[view_index]['view'] > 5):
            e[0] += 1
        else:
            e[1] += 1
    
    # 购物车事件
    for cart_index in group_df[group_df.cart > 0].index:
        if not cart_index.freq:
            cart_index = pd.DatetimeIndex([cart_index], freq='3D')[0]
        if (group_df.ix[cart_index:cart_index + 2].buy.sum() == 0):
            e[5] += 1

    # 收藏事件
    if (len(group_df[group_df.favor > 0].index) > 0):
        e[6] += 1

    # 存储每个用户的物品-事件矩阵
    try:
        b_event_mt = ub_event_map.Get(index[0])
    except IndexError:
        ub_event_map.Ins(index[0], sp.sparse.lil_matrix((len(bid_map), len(e)), dtype='int32'))
        b_event_mt = ub_event_map.get(index[0])
    b_event_mt[index[1], :] = e
    
    # 累加每个物品的事件向量
    try:
        b_event = b_event_map.Get(index[1])
    except IndexError:
        b_event_map.Ins(index[1], np.zeros((1, 7), dtype='int32')[0])
        b_event = b_event_map.get(index[1], np.zeros((1, 7), dtype='int32')[0])
    b_event += e
    #event_map.Ins(index, e)

linreg_dataset = pd.concat(linreg_dataset_list, ignore_index=True)
logreg_dataset = pd.concat(logreg_dataset_list, ignore_index=True)
linreg_dataset.to_csv("data/season1/train_linreg.csv", encoding="gbk", index=False)
logreg_dataset.to_csv("data/season1/train_logreg.csv", encoding="gbk", index=False)
```

```python
import pycuda.autoinit
import pycuda.gpuarray as gpuarray
import scikits.cuda.linalg as linalg
import scikits.cuda.misc as misc
from sklearn.preprocessing import normalize

# 计算每个用户的事件-偏好变换矩阵，算出偏好向量并求出用户对物品的打分
rate_mt = lil_matrix((len(uid_map), len(bid_map)), dtype='float32')
linalg.init()
for uid, b_event_mt in ub_event_map.iteritems():
    b_event_gpu = gpuarray.to_gpu(np.array(b_event_mt.toarray(), dtype='float32'))
    ones_gpu = misc.ones((b_event_mt.shape[0], 1), dtype='float32')
    sum_e_gpu = linalg.dot(b_event_gpu, ones_gpu, transa='T')
#    print np.allclose(np.array(b_event_mt.sum(axis=0)).transpose(), sum_e_gpu.get())
    sum_e = sum_e_gpu.get().transpose()[0]
    part_sum_e = sum_e[:-1].reshape(3, 2).sum(axis=1)
#    sum_e = np.array(b_event_mt.sum(axis=0))
#    part_sum_e = sum_e[0][:-1].reshape(3, 2).sum(axis=1)

    # 从未接触过的物品数量(收藏事件的变换参数)
    ones_gpu = misc.ones((b_event_mt.shape[1], 1), dtype='float32')
    action_gpu = linalg.dot(b_event_gpu, ones_gpu)
    untouched = len(action_gpu.get().nonzero()[0])
#    print np.allclose(sum([1 for i in xrange(b_event_mt.shape[0]) if len(b_event_mt[i, :6].nonzero()[0]) > 0]), untouched)
#    untouched = sum([1 for i in xrange(b_event_mt.shape[0]) if len(b_event_mt[i, :6].nonzero()[0]) > 0])
    
    # 事件-偏好变换矩阵(3*7)*(N*7)'
    csr_row = np.array([0, 0, 1, 1, 1, 2, 2], dtype='int32')
    csr_col = np.array([0, 1, 2, 3, 6, 4, 5], dtype='int32')
    csr_data = np.array([max(sum_e[1], 1), max(sum_e[0], 1),
                                     max(sum_e[3], 1), -max(sum_e[2], 1),
                                     (b_event_mt.shape[0] - untouched) / max(sum_e[6], 1), 
                                     max(sum_e[5], 1), max(sum_e[4], 1)], dtype='float32')
#    csr_data = np.array([sum_e[0, 1] / max(part_sum_e[0], 1), sum_e[0, 0] / max(part_sum_e[0], 1),
#                                     sum_e[0, 3] / max(part_sum_e[1], 1), -sum_e[0, 2] / max(part_sum_e[1], 1),
#                                     sum_e[0, 6] / max(b_event_mt.shape[0] - untouched, 1), 
#                                     sum_e[0, 4] / max(part_sum_e[2], 1), sum_e[0, 5] / max(part_sum_e[2], 1)], dtype='float32')
    param_mt = csr_matrix((csr_data, (csr_row, csr_col)), shape=(3, 7), dtype='float32')
    
    # 偏好矩阵(3*N)
    param_gpu = gpuarray.to_gpu(np.array(param_mt.toarray(), dtype='float32'))
    profile = linalg.dot(param_gpu, b_event_gpu, transb='T')
#    profile = param_mt.dot(b_event_mt.transpose().tocsr())
#    profile.eliminate_zeros()
    
   # 归一化偏好向量的每个分量
    profile = gpuarray.to_gpu(normalize(profile.get(), norm='l2', axis=1))
    
    # 评分矩阵
    weight_gpu = gpuarray.to_gpu(np.array([0.3, 0.6, 0.1], dtype='float32'))
    rate_mt[uid, :] = linalg.dot(weight_gpu, profile).get()
#    print np.allclose(linalg.dot(weight_gpu, profile).get(), csr_matrix(np.array([0.3, 0.6, 0.1])).dot(profile1).toarray())
#    rate_mt[uid, :] = csr_matrix(np.array([0.3, 0.6, 0.1])).dot(profile)
```
